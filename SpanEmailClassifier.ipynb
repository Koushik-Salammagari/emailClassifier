```python
import pandas as pd
df = pd.read_csv('spam.csv', encoding='latin-1')
#df = pd.read_csv('spam.csv')
df.head()
```

```python
df = df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis='columns')
df.head()

# once done cannot be changed
```

```python
df.head()
```

```python
df = df.rename(columns={'v1': 'Category', 'v2': 'Message'})
```

```python
df.groupby('Category').describe()
```

```python
# the ham count is 4825
# the spam count is 747
```

```python
# Binary classification
# if is a spam them return 1
#if not return 0 ( no worries )
```

```python
# creating a new column of 0's and 1's
df['spam'] = df['Category'].apply(lambda x: 1 if x=='spam' else 0)
```

```python
df.head()
```

```python
# calling sklearn .....importigng test-train split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.25)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
# making use of count vectorizer
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:2] # checking the first 2 rows
```

```python
# Naive Bayes Classification
# Mutinomial NB
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
emails = ['Is that seriously how you spell his name?',
          'SMS. ac Sptv: The New Jersey Devils and the Detroit Red Wings play Ice Hockey. Correct or Incorrect? End? Reply END SPTV']

emails_vector = v.transform(emails)
model.predict(emails_vector)
```

```python
# Classified correctly
```

```python
# checking for the accuracy score
X_test_count = v.transform(X_test)
model.score(X_test_count, y_test)
```

```python
# creating sk-learn pipeline
from sklearn.pipeline import Pipeline
classifier = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
classifier.fit(X_train, y_train)
```

```python
classifier.score(X_test, y_test)
```
